```python
%run nn_functions.ipynb
tf.reset_default_graph()
from tensorflow.examples.tutorials.mnist import input_data
#录入数据
mnist=input_data.read_data_sets('MNIST_data',one_hot=True)
batchsize=50
nsize=mnist.train.num_examples//batchsize
columnsize=28#图片像素的列数，因为是逐行传入inputsize
linesize=28#图片像素的行数，maxinputsize
lstmsize=100#lstm核的大小
nclass=10#分类大小

#创建占位符
x=tf.placeholder(tf.float32,[None,784],name='xinput')
y=tf.placeholder(tf.float32,[None,10],name='yinput')

#初始化权值和偏置值
w=weight_variable([lstmsize,nclass])
b=bias_variable([nclass])

#定义rnn函数，其中对mnist数据集的数据形式重构为矩阵形式，若数据集形式变化可修改x_4d
def RNN(x,w,b):
    #重构mnist数据集的数据为矩阵形式
    x_4d=tf.reshape(x,[-1,linesize,columnsize])
    lstmcell=tf.contrib.rnn.BasicLSTMCell(lstmsize)
    outputs,fs=tf.nn.dynamic_rnn(lstmcell,x_4d,dtype=tf.float32)#这里dynamic__rnn 可以理解为一个矩阵相乘的运算
    result=tf.nn.softmax(tf.matmul(fs[1],w)+b)
    return result,outputs

#计算rnn网络的预测值
prediction,outputs=RNN(x,w,b)

#损失函数
loss=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=y,logits=prediction))
#训练目标及优化器选取
Trainstep=tf.train.AdamOptimizer(1e-4).minimize(loss)


#准确率
correctprediction=tf.equal(tf.argmax(y,1),tf.argmax(prediction,1))
accuracy=tf.reduce_mean(tf.cast(correctprediction,tf.float32))

#存储训练模型
saver=tf.train.Saver()
#训练神经网络
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(31):
        for _ in range(nsize):
            bx,by=mnist.train.next_batch(batchsize)
            sess.run(Trainstep,feed_dict={x:bx,y:by})
            
        acc=sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels})
        op=sess.run(outputs,feed_dict={x:mnist.test.images,y:mnist.test.labels})
        print("accuracy="+str(acc)+"outputs="+str(outputs))
    saver.save(sess,'train_models/rnn_trained.ckpt')
    #之后调用模型用语句tf.train.Saver().restore(sess,'文件夹的ckpt文件')即可
        
```
